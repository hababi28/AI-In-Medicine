### Introduction to Lists

A list is a versatile data type in Python, which can hold an ordered sequence of elements. The syntax to define a list involves using square brackets [], with each element separated by a comma.

Characteristics:
- Lists are ordered, meaning the elements have a specific arrangement that is maintained.
- Lists are mutable, meaning you can change their content after they are created.
- Elements in a list can be of mixed types, including strings, integers, floats, and even other lists.

````python
my_list = [1, 'apple', 3.14, [2, 3, 4]]
````

In this example, my_list contains an integer (1), a string ('apple'), a float (3.14), and another list ([2, 3, 4]).

#### Slicing
Get a subset of the list.

``sublist = my_list[start:end]``

**start**: The index of the first element you want in the slice. The slice starts with the element at this index.

**end**: The index of the first element you don't want in the slice. The slice goes up to, but does not include, the element at this index.


If start is omitted, it defaults to 0, which means the slice will start from the first element.

If end is omitted, the slice will go up to and include the last element of the list.

Negative indices can be used for start and end, in which case counting starts from the end of the list. For example, -1 refers to the last element, -2 to the second last, and so on.


````python
sublist = my_list[1:3]

print(sublist) # Returns the second and third elements of the list. Returns [1,2] out of [0,1,2,3].
````

````python
sublist = my_list[:2]  # Result: [0, 1, 2, 3]
print(sublist) # Returns that first two elements of the list. Returns [0,1] from [0,1,2,3].
````

````python
sublist = my_list[2:]  # Result: [0, 1, 2, 3]
print(sublist) # Returns that last two elements of the list. Returns [2,3] from [0,1,2,3].
````

A set in Python is a built-in data type that holds an unordered collection of unique elements. This means that any duplicates are automatically removed.

**Note:** Sets don't preserve the order in which elements were added, and you can't access elements by index like you can with lists. So sets are a good choice when you want to store a collection of unique elements and don't care about their order or being able to access them by index.

Below we have a function that generates the fibonacci sequence of length `n`.

**What is a function?**

In Python, a function is a reusable piece of code. Functions are useful for tasks that you'll perform more than once.

A function in Python is defined by the "def" keyword, followed by the function's name,
a pair of parentheses containing any parameters the function may take, and a colon.
The body of the function is indented under the definition.

We will cover functions in more detail in week 2.

````python
def generate_fibonacci(n):
    a, b = 0, 1
    fib_sequence = []

    for _ in range(n):
        fib_sequence.append(a)
        a, b = b, a + b

    return fib_sequence
generate_fibonacci(20)
````

## Matrices

A **matrix** is a two-dimensional array of elements, commonly represented as a list of lists in Python. Each inner list represents a row of the matrix. You can access individual elements by using two indices: the first for the row and the second for the column.

Below is a quick example that demonstrates how to create a matrix, access its elements, and iterate over all its values.


````python
# Define a 3x3 matrix using a list of lists
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Access a specific element:
# For example, to access the element in the second row, third column:
element = matrix[1][2]  # This will be 6, as indexing starts at 0
print("Element at row 2, column 3:", element)

# Iterate over the matrix and print each element
print("Matrix elements:")
for row in matrix:
    for value in row:
        print(value, end=" ")
    print()  # Newline after each row
````
