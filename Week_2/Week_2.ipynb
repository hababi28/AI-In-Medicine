# Introduction to Functions

In Python, a function is a reusable piece of code. Functions are useful for tasks that you'll perform more than once.

A function in Python is defined by the `def` keyword, followed by the function's name,
a pair of parentheses containing any parameters the function may take, and a colon.
The body of the function is indented under the definition.


## Examples

```python
for i in range(3):
    print('Hello, World!')
```

```python
for i in range(3):
    print(i)
```

```python
def print_hello_world(n):
    for _ in range(n):
        print("Hello, World!")

# Call the function
print_hello_world(3)
```

```python
# Let's start with a very simple function:

# This function takes no arguments and prints a greeting
def greet():
    print("Hello, World!")

# We can call the function like this:
greet()
```

```python
# Functions can also take arguments:

# This function takes one argument and greets the person by name
def greet_person(name):
    print(f"Hello, {name}!")

# We call this function by passing an argument:
greet_person("Alice")
```

```python
# Functions can also return values. This is done using the "return" keyword.

# This function adds two numbers and returns the result
def add_numbers(num1, num2):
    sum = num1 + num2
    return sum

# We can call this function and store its return value in a variable:
result = add_numbers(2, 3)
print(result)  # prints: 5
```

# Regular Expressions

After learning how to interact with files, it's useful to know about regular expressions. Regular expressions, often abbreviated as regex, they are patterns that specify text searches. These are incredibly useful for various tasks, such as searching, splitting, and replacing text in files.

**What is a Regular Expression?**

A regular expression is a sequence of characters that forms a search pattern. For example, you might use the regex ``\d{3}-\d{2}-\d{4}`` to search for occurrences of a pattern in the format "123-45-6789" (like a Social Security Number) within a text file.


---


**What does '\d{3}-\d{2}-\d{4}' mean?**

*   \d - digit
*   {n} - The number of digits e.g. 123 has 3 digits, 45 has 2 digits and 6789 has 4 digits.
* we put the hyphen '-' in places where we expect one in that position of text. We can do this with letters and numbers too, if we know they will be in that position beforehand.

**Python re Module**

Python has a built-in package called ``re`` that can be used to work with regular expressions.

**Basic Regex Functions**
- ``re.findall()``: Returns all non-overlapping matches in a string as a list.
- ``re.search()``: Searches the string for a match and returns a match object if found.
- ``re.split()``: Splits the source string by the occurrences of the pattern.
- ``re.sub()``: Replaces occurrences of the pattern with another string.



## Examples

Here we have an abstract of a medical paper, let's try and extract some useful information.

```python
import re

with open('medical_paper.txt', 'r') as file:
    data = file.read()

print(data)
```

```python
# Find All UK Phone Numbers in a Text File
# The phone numbers are in the format +44 00 0000 0000.

pattern = r"\+44 \d{2} \d{4} \d{4}"
result = re.findall(pattern, data)
print("Found Phone Numbers:", result)
```

Real-world data often contains imperfections, including spelling errors, which can significantly impact data analysis and outcomes. An example of such an error is the misspelling of the word "Aspirin" as "Aspiriin" in the text file 'medical_paper.txt'. Cleaning and correcting data is a crucial step to ensure its quality and reliability. Here is an example of how you would do that:

```python
# Replace All Occurrences of a Specific Word
with open('medical_paper.txt', 'r') as file:
    data = file.read()

new_data = re.sub(r'\bAspiriin\b', 'Aspirin', data)
print(new_data)

# Writing the modified content back to the file
with open('medical_paper_corrected.txt', 'w') as file:
    file.write(new_data)
```

```python
# Find All Capitalised Words
pattern = r"\b[A-Z][a-z]*\b"
result = re.findall(pattern, new_data)
print("Found Capitalised Words:", result)
```

|---------|-----------------------------------------------------------|------------------|----------------|

| `{m}`   | Exactly m repetitions of preceding element                 | `a{3}`           | `aaa`          |
| `{m,n}` | From m to n repetitions of preceding element               | `a{1,3}`         | `a`, `aa`, `aaa` |

| `|`     | Either/or                                                  | `a|b`            | `a`, `b`       |
| `()`    | Grouping                                                   | `(abc){2}`       | `abcabc`       |
| `\`     | Escape special characters                                  | `\.`             | `.`            |
| `\d`    | Any digit                                                  | `\d{2}`          | `09`, `11`     |
| `\b`    | Empty string at beginning/end of word                      | `\babc\b`        | `abc`          |
| `\w`    | Any alphanumeric character                                 | `\w+`            | `abc123`       |

# Introduction to Classes:


In Python, a class serves as a blueprint for creating objects. Objects themselves are individual instances of a class and contain both data (with its various attributes) and functionalities (through its methods).

**Data (Attributes)**

Attributes are variables that belong to the class. They hold information that is relevant to the object. For example, in a ``Patient`` class, attributes could include ``name``, ``age``, and ``medical_history``.

**Actions (Methods)**

Methods are functions that belong to the class and can be called on any object of that class. They usually manipulate the attributes or perform specific tasks. In a ``Patient`` class, methods could include ``add_diagnosis()`` or ``schedule_appointment()``.

Classes are used in a style of programming called **Object Orientated Programming** (OOP).

What is **Object Orientated Programming** (OOP)?

Object-Oriented Programming (OOP) is a programming paradigm based on the concept of "objects," which represent both data and the functionalities to manipulate that data. In OOP, code is organised into reusable "classes" and "objects" to make it more modular, maintainable, and intuitive to understand.

**Core Concepts of OOP**
- Encapsulation: The bundling of data (attributes) and methods (functionalities) into a single unit (object).
- Inheritance: The mechanism by which one class can inherit the attributes and methods from another class.
- Polymorphism: The ability to present the same interface for different data types.
- Abstraction: Hiding complex reality while exposing only the necessary parts.

What is **Procedural Programming**?

OOP is an alternative to **Procedural Programming**. **Procedural Programming** is focussed on just writing functions to operate on data.


|                        | Procedural Programming        | Object-Oriented Programming |
|------------------------|------------------------------|----------------------------|
| **Advantages**         |                              |                            |
| Simplicity             | Easier to understand and write, especially for small tasks. | Better modularity makes it easier to manage complex systems. |
| Performance            | May execute faster due to lower overhead. | -                          |
|                        |                              |                            |
| **Disadvantages**      |                              |                            |
| Scalability            | Harder to scale and maintain as projects grow. | -                        |
| Code Reuse             | Limited code reuse capabilities. | Inheritance allows for significant code reuse. |
| Complexity             | -                            | Can be complex for simple projects. |
| Performance            | -                            | Potential overhead of method calls and attribute look-ups. |


When to Use Which?

- Small, Simple Projects: Procedural programming is often sufficient.
- Large, Complex Projects: OOP offers greater maintainability and is more scalable.
- Team-based Development: OOP is generally more structured and easier for multiple developers to collaborate on.

# The ``Random`` Module

The random module in Python is used to generate pseudo-random numbers for various probabilistic distributions. It is an extremely useful module for tasks such as simulating events, generating random data, and for games and gambling mechanics. The numbers generated with random are not truly random; they are deterministic, but are usually sufficient for most cases where unpredictable results are desired.

## Examples

```python
# Let's first import the module
import random
```

### Random Float: ``.random()``

```python
print(random.random()) # Output: float between 0.0 and 1.0
```

### Random Integer: ``.randint()``

```python
print(random.randint(1, 10))  # Output: integer between 1 and 10
```

### Random Element from a List: ``.choice(list)``

```python
names = ['Alice', 'Bob', 'Charlie', 'Diana']
print(random.choice(names))  # Output: randomly chosen name from the list
```

### Shuffle a List: ``.shuffle(list)``

```python
numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print(numbers)  # Output: the list 'numbers' is shuffled
```

# Text-Based Game

In this text-based game, players battle with unique characters, each possessing distinct health, moves, and catch rates. The goals? Defeat opponents and capture characters by making smart moves and timing your captures perfectly.

Each character is defined with a specific set of attributes and abilities, as outlined in the `Character` class.



## Starting Code

We provide initial code that includes the `Character` class, attack functions, and capture mechanics. Your task is to enhance the game by adding new features to the game by writing your own code as part of the exercises below.


```python
# Import
import random

# Define the character class
class Character:
    def __init__(self, name, health, moves, catch_rate):
        self.name = name
        self.health = health
        self.moves = moves  # This is a dictionary of moves with their hit percentages
        self.catch_rate = catch_rate  # Catch rate would depend on rarity and current health

    def attack(self, move, opponent):
        if random.random() <= self.moves[move]:
            damage = random.randint(10, 20)  # Random damage between 10 and 20
            opponent.health -= damage
            print(f"{self.name} used {move}. It's a hit! {opponent.name} lost {damage} health!")
        else:
            print(f"{self.name} used {move}. But it missed!")
```

```python
def throw_ball(character):
    if random.random() <= character.catch_rate:
        print(f"Success! You caught {character.name}!")
    else:
        print(f"Oops! {character.name} escaped!")

```

```python
# Create characters
char1 = Character("Char1", 100, {"WeaponA": 0.85, "WeaponB": 0.95}, 0.3)
char2 = Character("Char2", 100, {"WeaponC": 0.85, "WeaponD": 0.95}, 0.3)

# Simulate Char1 attacking Char2
char1.attack("WeaponA", char2)

# Try to catch Char1
throw_ball(char2)
```
