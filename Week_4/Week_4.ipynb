## What is Machine Learning (ML)?

"Machine learning is a branch of artificial intelligence (AI) and computer science which focuses on the use of data and algorithms to imitate the way that humans learn, gradually improving its accuracy." - IBM.com


In other words, machine learning technology uses an algorithm to teach the computer how to solve problems and gain insights from solving those problems. That’s how the computer learns automatically, without human intervention or assistance: by observing and looking for patterns in data and using feedback loops to monitor and improve its predictions. While humans would be overwhelmed with masses of data, machine learning thrives and is able to evolve its understanding in order to make better decisions in the future, based on the examples that were provided to it.


There are three main types of machine learning:


- Supervised Learning - Training a model on labeled data to predict outcomes based on input. Used in regression, classification, and forecasting.

- Unsupervised Learning - The model identifies patterns and structures in unlabeled data. Commonly used for clustering, association, and dimensionality reduction.

- Reinforcement Learning - An agent learns to make decisions in an environment by optimizing rewards. Used in gaming, robotics, and real-time decision-making.

We will focus on Supervised learning for now.

## What is Deep Learning?

Deep Learning is a specialised subset of Machine Learning.

This involves neural networks with many layers, hence the term "deep". If a network has more than 3 total layers (input, hidden and output), so more than 1 hidden layer, it is "deep". These neural networks attempt to simulate the behavior of the human brain, but in a simplified form, in order to process data and create patterns for use in decision making.

Deep Learning models automatically learn to represent data through multiple layers of abstraction, enabling them to make complex decisions based on large amounts of unstructured data. In other words, each layer of the neural network processes the input data a little bit more, extracting increasingly sophisticated features. The first layer might learn to recognize basic patterns, like edges in an image. The next layer uses this information to recognize more complex shapes, like textures or corners. Subsequent layers build upon this, eventually recognizing highly complex features, such as faces in an image or the sentiment in a piece of text.

These models are particularly powerful in fields such as image and speech recognition, natural language processing, and autonomous vehicles, where they can learn from vast quantities of data.

### Classification vs. Regression in Machine Learning

Two very common tasks in Machine Learning are classification and regression. These two types of tasks differ in their goals, output, and the types of problems they address, therefore Understanding the difference between classification and regression is vital when choosing the right machine learning approach for a specific problem, as it greatly influences model selection, training, and evaluation.

**Classification:**

- **Goal:** In classification tasks, the primary goal is to categorize input data points into predefined classes or categories. The aim is to assign a label or class to each data point based on its features or attributes.

- **Output:** The output of a classification model is a discrete label or category. It answers questions like "Is this email spam or not?" or "Is this an image of a cat, a dog, or something else?"

- **Examples:** Examples of classification tasks include spam email detection (binary classification), image classification (multiclass classification), and sentiment analysis (multiclass or binary classification).

- **Evaluation Metrics:** Common evaluation metrics for classification tasks include accuracy, precision, recall, F1-score, and ROC-AUC.

**Regression:**

- **Goal:** In regression tasks, the goal is to predict a continuous numerical value or quantity. Regression models aim to establish a relationship between input features and a numeric target variable.

- **Output:** The output of a regression model is a numerical value. It answers questions like "What will be the house price based on its features?" or "What will be the temperature tomorrow?"

- **Examples:** Examples of regression tasks include predicting house prices, estimating sales revenue, and forecasting stock prices.

- **Evaluation Metrics:** Common evaluation metrics for regression tasks include Mean Absolute Error (MAE), Mean Squared Error (MSE), Root Mean Squared Error (RMSE), R-squared (R²), and Mean Absolute Percentage Error (MAPE).

**Key Differences:**

1. **Output Type:** The fundamental difference between classification and regression is the type of output they produce. Classification outputs discrete categories, while regression outputs continuous numerical values.

2. **Use Cases:** Classification is used when you want to assign labels or categories to data points, often for decision-making or sorting tasks. Regression is used when you want to predict a quantity or value, which is typically used for estimation or forecasting.

3. **Evaluation Metrics:** Each type has its own set of evaluation metrics tailored to its output type. Classification uses metrics like accuracy, precision, recall, and F1-score, while regression uses metrics like MAE, MSE, RMSE, and R-squared.

#### Load Data  

- In machine learning, the first step is to acquire and load the data that will be used for training and testing your model. This data typically comes in the form of structured datasets that contain both input features (variables) and the target variable (what you want the model to predict).

- Loading the data involves importing it from various sources such as databases, CSV files, APIs, or web scraping.
- It's essential to understand your data, explore it, and preprocess it to handle missing values, outliers, and other data quality issues.

#### Process Data

- Data preprocessing is a crucial step in machine learning and often consumes a significant amount of time. It involves cleaning and transforming the data to make it suitable for training a machine learning model.

- Common data preprocessing tasks include:
   - Handling missing data: Deciding how to fill or impute missing values.
   - Feature scaling: Ensuring all features have similar scales (e.g., using normalization or standardization).
   - Encoding categorical variables: Converting categorical data into numerical form (e.g., one-hot encoding).
   - Feature selection: Choosing the most relevant features to improve model performance and reduce overfitting.
   - Data splitting: Splitting the data into training and testing sets allows us to evaluate how well the model generalises to new, previously unseen data. In short, it simulates how the model will perform when deployed in the real world.


#### Create Model

- Once the data is prepared, you design the architecture of your machine learning model. The choice of model depends on the problem you're trying to solve (e.g., classification, regression) and the nature of your data.

- Creating the model involves selecting the type of algorithm (e.g., decision trees, neural networks, support vector machines) and defining its structure, including the number of layers, neurons, and activation functions.
- You'll also configure various hyperparameters, which are settings that control how the model learns, such as learning rate, batch size, and regularization strength.


#### Train Model

- Model training is the process of feeding the prepared data into the model and adjusting its internal parameters to make predictions as accurate as possible.

- During training, the model learns patterns and relationships within the data through iterative optimization techniques like gradient descent.
- The training process aims to minimize a defined loss or cost function, which quantifies the difference between the model's predictions and the actual target values.
- The duration of training varies depending on the complexity of the model and the size of the dataset.


#### Evaluate Model

- After training, it's essential to assess the model's performance to understand how well it generalises to new, unseen data. Different metrics are used based on the kind of task (such as Classification or Regression). Please find some common evaluation metrics in the table below:

**Classification Metrics**


| Metric              | Description                                     |
|---------------------|-------------------------------------------------|
| Accuracy            | Measures overall correctness of classification. |
| Precision           | Measures the proportion of true positives among all predicted positives. |
| Recall (Sensitivity)| Measures the proportion of true positives among all actual positives. |
| F1-Score            | Harmonic mean of precision and recall, balances precision and recall. |
| ROC-AUC             | Measures the area under the Receiver Operating Characteristic curve. |
| Log Loss            | Measures the quality of probabilistic predictions. |

**Regression Metrics**


| Metric              | Description                                     |
|---------------------|-------------------------------------------------|

| Root Mean Squared Error (RMSE) | Represents the square root of MSE, providing the error in the original unit. |
| R-squared (R²)     | Measures the proportion of the variance explained by the regression model. |
| Mean Absolute Percentage Error (MAPE) | Expresses the error as a percentage of the actual values. |

- Evaluating the model helps identify if it is overfitting (performing well on the training data but poorly on test data) or underfitting (not capturing the underlying patterns).
- The evaluation results guide further model improvement steps.



#### Hyperparameter Tuning

- Tuning the model's hyperparameters involves systematically adjusting settings to find the best combination for optimal performance. This is often done using techniques like grid search or random search.




#### Feature Engineering

- Feature engineering entails creating new features or modifying existing ones to improve the model's ability to make accurate predictions. This can involve domain knowledge and creativity.
- After making adjustments, you retrain and evaluate the model to see if it has improved. This iterative process may be repeated until satisfactory results are achieved.

### Linear regression

Linear regression is a simple way to spot patterns in data by drawing a straight line through a set of points. Imagine you’ve got dots on a graph, each dot shows two numbers, like hours studied and test scores. Linear regression finds the best line that cuts through them, with about the same number of dots above it as below it.

We use this line to understand how the numbers relate and to guess what might happen with new data.

image.png

#### Example: Iris dataset

We have iris plants of three species and we have 50 samples in total.  

```python
# Import what we need
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Load the iris dataset
iris = load_iris()
X = iris.data[:, 2].reshape(-1, 1)  # Petal length (input)
y = iris.data[:, 3]  # Petal width (what we're predicting)

# Create and train our model
model = LinearRegression()
model.fit(X, y)

# Get coefficient and intercept (our formula)
m = model.coef_[0]
b = model.intercept_
print("Y = mx + c")
print(f"Formula: Width = {m:.2f} × Length + {b:.2f}")

# Calculate how well our model fits the data
score = model.score(X, y)
print(f"R² score: {score:.2f}")

# Create a scatter plot with prediction line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, c=iris.target, cmap='viridis')
plt.plot(X, model.predict(X), color='red', linewidth=2)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Predicting Iris Petal Width from Length')
plt.colorbar(label='Species')
plt.tight_layout()
plt.show()
```

### Logistic Regression (used for Classification)

Now that we've seen how linear regression works for predicting continuous values, let's explore logistic regression, which addresses a different type of problem.

While linear regression helps us answer questions like "How much?" or "How many?", logistic regression answers questions like "Which category?" or "Yes or No?".

### Key Differences:

- **Purpose**: Linear regression predicts continuous values; logistic regression classifies data into categories
- **Output**: Linear regression outputs any number; logistic regression outputs probabilities between 0 and 1
- **Transformation**: Logistic regression applies a sigmoid function to the linear equation, creating an S-shaped curve that naturally limits output to the range [0,1]

image.png

- **Interpretation**: The output represents the probability of belonging to a particular class

In our next example, we'll use logistic regression with the same Iris dataset, but instead of predicting measurements, we'll predict which species a flower belongs to based on its characteristics.

```python
# Import libraries
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
iris = load_iris()
X = iris.data[:, :2]  # Sepal length and width only
y = iris.target       # Species (0, 1, or 2)

# Train a logistic regression model
model = LogisticRegression(max_iter=200)
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)
accuracy = accuracy_score(y, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print("\nClassification Report:")
print(classification_report(y, y_pred, target_names=iris.target_names))

# Create a colorful decision boundary plot
def plot_decision_boundary(X, y, model, iris):
    # Set up the meshgrid
    h = .02  # Step size
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Make predictions on the meshgrid
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the decision boundary
    plt.figure(figsize=(10, 6))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')

    # Plot the training points
    scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')
    plt.xlabel('Sepal Length (cm)')
    plt.ylabel('Sepal Width (cm)')
    plt.title('Logistic Regression Decision Boundaries')

    # Fixed legend code - create a legend manually
    legend_elements = [plt.Line2D([0], [0], marker='o', color='w',
                                 markerfacecolor=plt.cm.viridis(i/2),
                                 markersize=10)
                      for i in range(3)]
    plt.legend(legend_elements, iris.target_names, title="Species")

    plt.tight_layout()
    plt.show()

# Show the decision boundaries
plot_decision_boundary(X, y, model, iris)
```

#### What's Happening Here?

Our logistic regression model is finding decision boundaries that separate the three iris species based on sepal measurements.

The plot shows:
- Each point is a flower, coloured by its actual species
- Colored regions are where the model predicts each species
- Boundary lines show where the model switches its prediction from one species to another

Unlike linear regression that fits a line to predict values, logistic regression creates boundaries that classify points into different categories.

#### Understanding Decision Boundaries

Those lines separating the coloured regions are our **decision boundaries**. For our 3-class problem, logistic regression creates **three separate formulas** - not just one. Each formula has the form:

`w₁ × sepal_length + w₂ × sepal_width + b = 0`

where w's are weights and b is bias.

The model actually creates three separate classifiers:
* "setosa vs not setosa"
* "versicolor vs not versicolor"
* "virginica vs not virginica"

When predicting, it runs a flower's measurements through all three formulas and picks the class with the **highest score**. The boundaries occur where two classes have equal scores - that's where the model switches from predicting one species to another.


### Decision tree

A decision tree is like a flowchart that asks simple yes-or-no questions to make decisions. Imagine you’re deciding what to do: “Am I hungry?” If yes, you ask, “Do I have $25?” If yes, you go to a restaurant; if no, you buy a hamburger. If you’re not hungry, you go to sleep. Each question splits the choices into branches until you reach a final answer.

We use decision trees to sort data step-by-step and predict outcomes for new situations.

|----------------------------|-------------------------------------------------------------------------------------------------|
| **Advantages**             | - Easy to understand and visualise, like a flowchart anyone can follow.                        |
|                            | - Works well with both numerical and categorical data, common in medical records.               |
|                            | - Handles missing data better than some methods, useful for incomplete health data.             |
| **Disadvantages**          | - Can get complex and hard to manage if there are too many branches (overfitting).              |
|                            | - May not perform well with very subtle patterns in data, like small changes in medical tests.  |
|                            | - Sensitive to small changes in data, which could affect medical predictions.                   |
| **When It’s Used (Especially in Medicine)** | - Diagnosing diseases by splitting patient data (e.g., symptoms, test results) into categories. |
|                            | - Predicting treatment outcomes, like which therapy might work best for a patient.              |
|                            | - Identifying risk factors for conditions, such as heart disease, using patient history.        |




```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Create our dataset
# Features: [hungry (0=No, 1=Yes), has_25_dollars (0=No, 1=Yes)]
# Decisions: 0=Sleep, 1=Restaurant, 2=Hamburger
X = np.array([
    [0, 0],  # Not hungry, no money -> Sleep
    [0, 1],  # Not hungry, has money -> Sleep
    [1, 0],  # Hungry, no money -> Hamburger
    [1, 1],  # Hungry, has money -> Restaurant
])

y = np.array([0, 0, 2, 1])  # The corresponding decisions

# Create and train the decision tree
# We set max_depth=2 to match our example tree structure
tree_clf = DecisionTreeClassifier(max_depth=2, random_state=42)
tree_clf.fit(X, y)

# Define feature and class names for better visualization
feature_names = ['Hungry?', 'Has $25?']
class_names = ['Sleep', 'Restaurant', 'Hamburger']

# Plot the tree
plt.figure(figsize=(12, 8))
plot_tree(tree_clf, feature_names=feature_names, class_names=class_names,
          filled=True, rounded=True, fontsize=12)
plt.title("Decision Tree: What Should I Do?")
plt.tight_layout()
plt.show()

# Print the decision rules
print("Decision Rules:")
print("1. If not hungry → Sleep")
print("2. If hungry and has $25 → Restaurant")
print("3. If hungry but doesn't have $25 → Hamburger")
```

```python
# Interactive testing of our decision tree
def test_decision_tree():
    print("\n===== Decision Tree Interactive Testing =====")
    print("Let's decide what you should do!")

    while True:
        try:
            # Get user input
            hungry = input("\nAre you hungry? (yes/no): ").lower()
            if hungry not in ['yes', 'no']:
                print("Please enter 'yes' or 'no'")
                continue

            if hungry == 'yes':
                has_money = input("Do you have $25? (yes/no): ").lower()
                if has_money not in ['yes', 'no']:
                    print("Please enter 'yes' or 'no'")
                    continue
            else:
                # If not hungry, we don't need to ask about money
                has_money = "doesn't matter"

            # Convert input to model format
            hungry_val = 1 if hungry == 'yes' else 0
            money_val = 1 if has_money == 'yes' else 0

            # Make prediction
            features = np.array([[hungry_val, money_val]])
            prediction = tree_clf.predict(features)[0]

            # Display result
            print("\nDecision Tree Recommends:", class_names[prediction])

            # Explanation
            if prediction == 0:
                print("You're not hungry, so you should go to sleep.")
            elif prediction == 1:
                print("You're hungry and have enough money, so you should go to a restaurant.")
            else:
                print("You're hungry but short on cash, so you should buy a hamburger.")

            # Ask if user wants to try again
            again = input("\nTry another scenario? (yes/no): ").lower()
            if again != 'yes':
                print("Thanks for testing the decision tree!")
                break

        except Exception as e:
            print(f"An error occurred: {e}")
            print("Let's try again.")

# Run the interactive testing function
test_decision_tree()
```

### Random forest

They are useful for predicting outcomes, such as test scores, flower types, or health results, especially with complex or messy data.

### XGBoost


XGBoost is an advanced version of decision trees that builds on them to make even better predictions. It uses a method called gradient boosting, where it creates a series of decision trees, each one correcting the mistakes of the previous ones by focusing on the errors and improving accuracy over time. It does this by looking at where the predictions went wrong, then using the “gradient”—a mathematical measure of how the error changes, showing both how far off the predictions are and in what direction—to guide the next tree to fix those specific mistakes. This process uses a “loss function,” which is a way to calculate the total error in the predictions, and “optimization,” which means adjusting the trees step-by-step to minimize that error, gradually getting closer to the right answers. It’s like a team of trees learning from each other, guided by precise math to improve.

XGBoost is useful for predicting outcomes, such as disease risks, flower classifications, or other complex patterns, especially with large or tricky datasets.

#### Forward and Backward Propogation

##### Forward Propogation:

Artificial Neural Networks (ANNs) are computing systems inspired by the brain's neural networks. They are based on a collection of connected units called artificial neurons, or "nodes". Each connection (like the synapses in a biological brain) can transmit a signal from one node to another. The receiving node processes the signal and signals downstream nodes.

In Forward Propagation, we pass the input data through the network, node by node, until we get the output.


image.png

https://cdn-images-1.medium.com/v2/resize:fit:800/1*T4ARzySpEQvEnr_9pc78pg.jpeg

To look more closely at the image, it shows the fundamental building block of neural networks - a single neuron. During forward propagation, data flows from left to right. First, each input (x) is multiplied by its corresponding weight (ω). These weighted inputs are summed together, and a bias term (b) is added - think of bias as the neuron's baseline tendency to fire. This sum then passes through an activation function (φ), which introduces non-linearity, determining whether and how strongly the neuron fires. The final output (y) can either be our prediction or become input to the next layer. This simple process - multiply, sum, activate - is the essence of how neural networks learn patterns in data.

**Activation Functions**

Activation functions in neural networks help decide whether a neuron should be activated or not. Here's a table of the main activation functions:

| Activation Function | Characteristics                              | Common Usage                 |
|---------------------|----------------------------------------------|------------------------------|
| ReLU (Rectified Linear Unit) | Non-linear, range from 0 to ∞ (infinity)   | Hidden layers in neural networks |
| Sigmoid             | S-shaped, range from 0 to 1                  | Binary classification (output layer) |
| Softmax             | Normalizes outputs to a probability distribution (0 to 1) | Multi-class classification (output layer) |
| Tanh (Hyperbolic Tangent) | S-shaped, range from -1 to 1             | Rarely used in hidden layers |

ReLU is frequently used in most layers except the output because it helps mitigate the vanishing gradient problem, allowing models to learn faster and perform better. It's computationally efficient and provides non-linearity, enabling the network to handle complex data.

For the output layer, the choice of activation function depends on the nature of the task. When looking at classification tasks:
- **Sigmoid**:
    - Ideal for binary classification tasks where there are only two classes, such as determining whether a tumor is benign or malignant.
    - Also used in multi-label classification, where each class is treated independently, allowing for multiple classes to be predicted simultaneously. For example, in a photo, identifying various features like 'sunny', 'outdoor', and 'portrait' independently.

- **Softmax**:
    - Used for multi-class classification tasks where classes are mutually exclusive, meaning an image can only belong to one class.
    -  For example, classifying an image as a 'cat', 'dog', or 'bird', where each image can only be one of these options.
    - Softmax outputs a probability distribution, meaning the probabilities for all possible classes sum up to 1, which is useful for determining the most likely single class for an instance.
    - Softmax is an extension of the sigmoid function, so when there are two classes (binary classification), the output will be the same as the sigmoid function.

    
    
In short, the Sigmoid function gives a probability curve for two classes, Softmax extends this concept to multiple classes, distributing the probabilities across each class in a way that they sum up to 1.

##### **Backpropogation**:

Backpropagation is a method used to calculate the gradient of the loss function with respect to the weights in the neural network. This is done by applying the chain rule to compute the derivative of the loss function with respect to the weights, thus allowing us to adjust the weights to minimize the loss, the process of adjusting the weights to minimize the loss is called Gradient Descent.

##### **Example**:
Here is an example of forward propogation, backpropogation and gradient descent in action:

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    # Activation function
    return 1 / (1 + np.exp(-x))

def forward_propagation(inputs, weights, bias):
    # Forward propagation function
    z = np.dot(inputs, weights) + bias
    output = sigmoid(z)
    return output

def sigmoid_derivative(x):
    # Derivative of sigmoid function
    return x * (1 - x)

def back_propagation(output, target_output, inputs):
    # Compute error
    error = target_output - output
    d_output = error * sigmoid_derivative(output)

    # Compute gradients
    dw = np.dot(inputs.T, d_output)
    db = np.sum(d_output, axis=0, keepdims=True)

    return dw, db

def gradient_descent(weights, bias, dw, db, learning_rate):
    # Adjust weights and bias
    weights += learning_rate * dw
    bias += learning_rate * db
    return weights, bias

# inputs = np.array([[0.5, 0.6, 0.1]]) # Input data with additional dimension
# weights = np.random.randn(3, 1) # Random weights
# bias = np.zeros((1, 1)) # Bias

# # Forward Propagation
# output = forward_propagation(inputs, weights, bias)
# print(f"Output: {output}")

# target_output = np.array([[0.8]]) # Target output with additional dimension
# learning_rate = 2 # the default learning rate in most ML frameworks is set to 0.01.


# # Backpropagation to compute gradients
# dw, db = back_propagation(output, target_output, inputs)

# # Gradient descent to update weights and bias
# weights, bias = gradient_descent(weights, bias, dw, db, learning_rate)

# print(f"Updated weights: {weights}")
# print(f"Updated bias: {bias}")

```

```python
inputs = np.array([[0.5, 0.6, 0.1]])
weights = np.random.randn(3, 1)
bias = np.zeros((1, 1))
target_output = np.array([[0.8]])
learning_rate = 1

# Initialise lists to store epochs, losses, and outputs
epochs = 30
losses = []
outputs = []

for epoch in range(epochs):
    # Forward Propagation
    output = forward_propagation(inputs, weights, bias)

    # Store the output and loss for plotting
    outputs.append(output[0][0])
    loss = np.mean(np.square(target_output - output))
    losses.append(loss)

    # Backpropagation and Gradient Descent
    dw, db = back_propagation(output, target_output, inputs)
    weights, bias = gradient_descent(weights, bias, dw, db, learning_rate)

# Plotting
fig, axs = plt.subplots(2, figsize=(10, 7))

# Output per Epoch with Data Points
axs[0].plot(range(epochs), outputs, 'b-', marker='o', markersize=5, markevery=1, label='Output per Epoch') # markevery=10 will only mark every 10th data point
axs[0].set_title('Output per Epoch')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Output')
axs[0].legend()

# Loss per Epoch with Data Points
axs[1].plot(range(epochs), losses, 'r-', marker='x', markersize=5, markevery=1, label='Loss per Epoch') # markevery=10 will only mark every 10th data point
axs[1].set_title('Loss per Epoch')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Loss')
axs[1].legend()

plt.tight_layout()
plt.show()

print(f"Final weights: {weights}")
print(f"Final bias: {bias}")
```

Here is an more detailed explantion of what is happening at each step:

1. Initialisation: We initialise inputs, weights, and bias. Inputs are the data we want to make a prediction for. Weights are randomly initialized and bias is initialized as zero.

    `inputs = np.array([[0.5, 0.6, 0.1]]) # Input data`

    `weights = np.random.randn(3, 1) # Random weights`

    `bias = np.zeros((1, 1)) # Bias`

2. Forward Propagation: We pass the inputs, weights, and bias into the `forward_propagation` function. This function computes the dot product of the inputs and weights, adds the bias, and then applies the sigmoid function to this result to get the output.

    `output = forward_propagation(inputs, weights, bias)`
    `print(f"Output: {output}")  # Output: [[0.41205319]]`

    The output here (0.41205319) is the prediction of our model given the current weights and bias.

3. Backpropagation: In backpropagation, we calculate how much the output would change if we changed the weights and biases a little bit. This is done by taking the derivative of the error with respect to the weights and bias (dw and db).

    `target_output = np.array([[0.8]]) # Target output`
    `dw, db = back_propagation(output, target_output, inputs)`

    The target_output (0.8) is the true value we want our model to predict. We compute dw and db which are gradients that tell us how much to change the weights and bias to reduce the error.
    
4. Gradient Descent: Gradient descent is the process of adjusting the weights and bias to minimize the error between the target_output and our model's output. We multiply our gradients by a learning_rate (0.01) to ensure that we don't make too large of an adjustment and then add these adjustments to the weights and bias.

    `weights, bias = gradient_descent(weights, bias, dw, db, learning_rate)`
    `print(f"Updated weights: {weights}")  # Updated weights: [[ 0.17212071] [-0.75926446] [0.14596986]]`
    `print(f"Updated bias: {bias}")  # Updated bias: [[0.00093986]]`

    The updated weights and bias are used for the next prediction, and this process is repeated many times to train the model.

    Please note that this is a single iteration of forward propagation, backpropagation, and gradient descent through the whole training data, this is called an epoch.
    
    In practice, you would perform many such epochs, additionally many iterations within an epoch if using stochastic gradient descent, and adjust the weights and bias each time to minimize the error.

##### Stochastic gradient descent - explanation

Stochastic gradient descent (SGD) is much more commonly used in practice, instead of using the entire dataset for each iteration like traditional Gradient Descent, SGD uses a single data point or small batch, making it faster and more memory-efficient.  Although this might result in more iterations within an epoch, SGD often takes fewer epochs to converge to a minimum loss value, especially with large datasets. By estimating the gradient using small subsets of data, SGD introduces the concept of "batches" and "epochs" which allow for frequent, varied updates to the model's parameters, often leading to better solutions, particularly for complex, non-convex problems.

```python
from sklearn.metrics import accuracy_score, log_loss, f1_score, precision_score, recall_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt


from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay


from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, log_loss, f1_score, precision_score, recall_score, confusion_matrix, roc_curve, auc



def evaluate_model(model, X_test, y_test):
    """
    Evaluate the model and return a pandas Series object containing various metrics.
    """


    # Assuming get_predictions_and_probabilities is defined elsewhere
    results = get_predictions_and_probabilities(model, X_train, X_test)
    y_pred_test = results['y_pred_test']
    y_proba_test = results['y_proba_test']

    # Calculate metrics
    accuracy_test = accuracy_score(y_test, y_pred_test)
    log_loss_test = log_loss(y_test, y_proba_test)
    f1_test = f1_score(y_test, y_pred_test)
    precision_test = precision_score(y_test, y_pred_test)
    recall_test = recall_score(y_test, y_pred_test)
    conf_matrix_test = confusion_matrix(y_test, y_pred_test)
    mse = mean_squared_error(y_test, y_pred_test)
    mae = mean_absolute_error(y_test, y_pred_test)
    rmse = mse ** 0.5
    r2 = r2_score(y_test, y_pred_test)



    # Print metrics
    print(f"\nTest Accuracy: {accuracy_test}")
    print(f"Test Log Loss: {log_loss_test}")
    print(f"Test F1 Score: {f1_test}")
    print(f"Test Precision: {precision_test}")
    print(f"Test Recall: {recall_test}")
    print(f"Test Confusion Matrix: \n{conf_matrix_test}")
    print(f"Test MSE: {mse}")
    print(f"Test MAE: {mae}")
    print(f"Test RMSE: {rmse}")
    print(f"Test R2: {r2}")


    # Plot ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_proba_test)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(4, 3))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc='lower right')
    plt.show()

    results = {
        'Model': type(model).__name__,
        'Accuracy': accuracy_test,
        'Log Loss': log_loss_test,
        'F1 Score': f1_test,
        'Precision': precision_test,
        'Recall': recall_test
    }
    return results
```
